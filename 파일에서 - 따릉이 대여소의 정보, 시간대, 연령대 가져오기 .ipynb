```python
file_path = '/content/drive/MyDrive/ddareung/2021/서울특별시 공공자전거 이용정보(시간대별)_21.01.csv'
```

```python
try:
    df = pd.read_csv(file_path, encoding='cp949')

    # 4. 데이터 처리 (수정됨)
    # '대여시간' 컬럼이 이미 시간대(0~23)를 나타내는 숫자이므로,
    # 복잡한 변환 없이 이 컬럼을 그대로 사용합니다.
    # 컬럼 이름을 '시간대'로 바꾸어 명확하게 해줍니다.
    df.rename(columns={'대여시간': '시간대'}, inplace=True)


    # 5. 연령대 및 시간대별 카운트
    print("--- 연령대별 이용 건수 ---")
    # '연령대코드' 컬럼의 값들을 카운트합니다.
    age_counts = df['연령대코드'].value_counts()
    print(age_counts)

    print("\n--- 시간대별 이용 건수 ---")
    # 새로 만든 '시간대' 컬럼의 값들을 카운트하고, 시간순으로 정렬합니다.
    time_counts = df['시간대'].value_counts().sort_index()
    print(time_counts)


except FileNotFoundError:
    print(f"오류: 파일 경로를 찾을 수 없습니다. '{file_path}'")
    print("file_path 변수에 올바른 CSV 파일 경로를 입력했는지 확인해주세요.")
except Exception as e:
    print(f"데이터 처리 중 오류가 발생했습니다: {e}")
```
